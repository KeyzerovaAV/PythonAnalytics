# Задача 1
1. Скачать данные по ссылке: https://www.kaggle.com/datasets/ionaskel/laptop-prices
2. Считать данные с помощью pandas
3. Вывести на экран первые 5 строк
4. Посмотреть на описание признаков и на их содержание

```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', sep = ',', encoding = 'windows-1251')
df.head()
```

## 1.1
Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU.

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
df['Cpu_Company'].value_counts()
```

## 1.2
Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя.

```python
def get_memory_amount(memory):
    memory_amount = memory.split(' ')[0]
    if memory_amount.endswith('GB'):
        memory_amount = float(memory_amount[:-2])
    elif memory_amount.endswith('TB'):
        memory_amount = float(memory_amount[:-2]) * 1024
    return memory_amount

df['Memory_Amount'] = df['Memory'].apply(get_memory_amount)
df.head()
```

```python
df['Memory_Amount'].value_counts()
```

## 1.3
Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.).


```python
df['Memory_Type'] = df['Memory'].apply(lambda x: x.split(' ')[1])
df.head()
```

```python
df['Memory_Type'].value_counts()
```

## 1.4
Удалите признаки Memory и ScreenResolution.

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
df.head()
```

```python
df.info()
```

# Задача 2

## 2.1
Создайте признак SSD, который изначально равен 0.

```python
df['SSD'] = 0
df.head()
```

## 2.2
Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD.

```python
df.loc[df['Memory_Type'] == 'SSD', 'SSD'] = 1
df.head()
```

## 2.3
Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный.

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float')
df.head()
```

```python
df['Weight'].dtype
```
