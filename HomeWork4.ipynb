## 1.2
Изучите стоимости недвижимости.

```python
plt.figure(figsize = (10, 5))
plt.hist(df['price'], bins = 50)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Значения стоимости')
plt.ylabel('Количество значений стоимости');
```

### Вывод:
По графику видим, что большее количество значений стоимости лежит в диапазоне от 200000 до 1000000. Также видим, что имеется очень много выбросов в сторону увеличения. Построим "ящик с усами" без выбросов.

```python
plt.figure(figsize = (5, 10))
plt.boxplot(df['price'], showfliers = False)
plt.title('Распределение стоимости недвижимости')
plt.xticks([1], ['Стоимость'])
plt.ylabel('Значения стоимости');
```

### Вывод:
Действительно, убрав выбросы, видим, что 25% квартиль = чуть более 300000, медиана = примерно 450000, 75% квартиль = примерно 650000.

# Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополните каждый график письменными выводами и наблюдениями.

```python
data4 = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 
            'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement']]
data4
```

```python
corr_matrix = data4.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot = True, linewidths = .5, cmap = 'coolwarm');
```

### Вывод:
По матрице корреляций и "тепловой карте" видим, что на стоимость недвижимости влияют следующие характеристики недвижимости: количество ванных комнат (средняя прямая связь), жилая площадь (высокая прямая связь), уровень конструкции и дизайна дома (заметная прямая связь), площадь дома, находящаяся выше земли (заметная прямая связь).

```python
plt.figure(figsize = (10, 5))
sns.lineplot(x = df['grade'], y = df['price'])
plt.title('Зависимость стоимости недвижимости от уровня конструкции и дизайна дома')
plt.xlabel('Уровень конструкции и дизайна дома')
plt.ylabel('Стоимость недвижимости');
```

### Вывод:
График подтверждает вывод по корреляционной матрице: чем выше уровень конструкции и дизайна дома, тем выше стоимость недвижимости.

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = df['sqft_above'], y = df['price'])
plt.title('Зависимость стоимости недвижимости от площади дома над землей')
plt.xlabel('Площадь дома над землей')
plt.ylabel('Стоимость недвижимости');
```

### Вывод:
По графику видим, что есть заметная прямая связь между площадью дома над землей и стоимостью недвижимости. При этом основная масса домов имеют площадь над землей до 4000 кв.м.

```python
plt.figure(figsize = (10, 5))
plt.scatter(df['sqft_above'], df['price'])
plt.title('Зависимость стоимости недвижимости от площади дома над землей')
plt.xlabel('Площадь дома над землей')
plt.ylabel('Стоимость недвижимости');
```

### Вывод:
То же самое видим и на графике Matplotlib: есть заметная прямая связь между площадью дома над землей и стоимостью недвижимости.

```python
sns.jointplot(x = df['sqft_living'], y = df['price'], kind = 'reg');
```

### Вывод:
На графике взаимного распределения видим, что имеется достаточно высокая прямая связь между жилой площадью и стоимостью недвижимости.

```python
sns.jointplot(x = df['sqft_living'], y = df['price'], kind = 'kde');
```

### Вывод:
То же самое видим на графике взаимного распределения с другим представлением: имеется достаточно высокая прямая связь между жилой площадью и стоимостью недвижимости. Также видим, что основная масса домов имеют жилую площадь не более 5000 кв.м, и имеются выбросы, как мы уже это видели на графиках распределения жилой квадратуры и стоимости недвижимости.
